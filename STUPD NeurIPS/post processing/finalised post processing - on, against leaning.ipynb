```python
import pandas as pd
from tqdm import tqdm
import os
import shutil
```

# copy images 

```python
# root_folder="E:\SPUR NEURIPS"
root_folder="C:/Users/Palaash.HPZ/Desktop/SPUR_NeurIPS"
preposition="original against leaning"
old_folder=os.path.join(root_folder,preposition)
new_folder=old_folder+"_new"

for category in tqdm(os.listdir(old_folder)):
    source_folder=os.path.join(old_folder,category)
    dest_folder=os.path.join(new_folder,category)
    rgb_="RGB"
    dest_folder=os.path.join(dest_folder,rgb_)
    folders=os.listdir(source_folder)
#     print(folders)
    image_folder=sorted(folders)[-1]
#     print(image_folder)
    image_folder=os.path.join(source_folder,image_folder)
#     print(image_folder)
    for file in sorted(os.listdir(image_folder),key=lambda x:int(x.split("_")[-1].split(".")[0]))[9::10]:
        file_path=os.path.join(source_folder,image_folder,file)
        dest_path=os.path.join(dest_folder,file)
#         print(file_path)
        if not os.path.exists(dest_folder):
            os.makedirs(dest_folder)
#         print(file_path)
#         print(dest_folder)
#         print(dest_path)
        shutil.copy2(file_path, dest_path)
#         break
#     break
```
